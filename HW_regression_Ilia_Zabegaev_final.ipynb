# 2. Описание датасета¶

**Описание каждого признака:**

- `Home Ownership` (Тип жилья): Показывает тип собственности на жилье, например, "Владение", "Аренда" или "Ипотека".
- `Annual Income` (Годовой доход): Отражает годовой доход заемщика в денежном выражении.
- `Years in current job` (Количество лет на текущей работе): Показывает количество лет, которое заемщик проработал на текущем месте работы.
- `Tax Liens` (Налоговые обременения): Отражает наличие налоговых обременений или задолженностей у заемщика.
- `Number of Open Accounts` (Количество открытых счетов): Указывает на общее количество открытых счетов у заемщика, включая кредитные карты, кредитные линии и другие.
- `Years of Credit History` (Количество лет кредитной истории): Показывает общую продолжительность кредитной истории заемщика в годах.
- `Maximum Open Credit` (Максимальная открытая кредитная линия): Отражает максимальную сумму кредитной линии, которая была предоставлена заемщику.
- `Number of Credit Problems` (Количество кредитных проблем): Указывает на количество проблем с кредитом, например, просроченных платежей или коллекций.
- `Months since last delinquent` (Количество месяцев с последней просрочки платежа): Показывает количество месяцев, прошедших с последней просрочки платежа заемщика.
- `Bankruptcies` (Банкротства): Отражает количество банкротств, которые заемщик может иметь в кредитной истории.
- `Purpose` (Цель кредита): Указывает на цель, для которой заемщик берет кредит, например, "автомобиль", "образование" или "недвижимость".
- `Term` (Срок кредита): Показывает срок кредита в месяцах, например, "краткосрочный" или "долгосрочный".
- `Current Loan Amount` (Текущая сумма кредита): Отражает текущую сумму, которую заемщик еще должен выплатить по кредиту.
- `Current Credit` Balance (Текущий кредитный баланс): Указывает на текущий остаток задолженности по кредитам и кредитным линиям.
- `Monthly Debt` (Ежемесячный долг): Показывает сумму ежемесячных платежей по всем задолженностям и кредитам.
- `Credit Score` (Кредитный рейтинг): Отражает кредитный рейтинг заемщика, который оценивает его платежеспособность и кредитный риск.


- `Credit Default` (Кредитный дефолт): **Целевая переменная**, которая указывает, возникнет ли у заемщика дефолт по кредиту (1 - дефолт, 0 - нет).

##Шаг 2 — Предобработка данных (Preprocessing)

* Обработать `выбросы`
* Обработать `пропуски` (медиана/мода для числовых, отдельная категория для строковых)
* Кодировать `категориальные признаки`
* Масштабировать `числовые признаки` (StandardScaler / MinMaxScaler) — важно для логистической регрессии

```python
#Исправление Credit Score (умноженные на 10 значения)
print("=== Credit Score до исправления ===")
print(train['Credit Score'].describe())

# Значения > 850 — скорее всего умножены на 10, делим их обратно
train['Credit Score'] = train['Credit Score'].apply(
    lambda x: x / 10 if x > 850 else x
)
test['Credit Score'] = test['Credit Score'].apply(
    lambda x: x / 10 if x > 850 else x
)

print("\n=== Credit Score после исправления ===")
print(train['Credit Score'].describe())
```

```python
#Замена значения-заглушки в Current Loan Amount
print(f"Строк с Current Loan Amount = 99999999: {(train['Current Loan Amount'] == 99999999).sum()}")

# Заменяем заглушку на NaN, затем заполним медианой
train['Current Loan Amount'] = train['Current Loan Amount'].replace(99999999, np.nan)
test['Current Loan Amount']  = test['Current Loan Amount'].replace(99999999, np.nan)

print(f"После замены — пропусков в Current Loan Amount (train): {train['Current Loan Amount'].isna().sum()}")
```

```python
#Обработка Years in current job (строка → число)
import re

def parse_years(val):
    if pd.isna(val):
        return np.nan
    if '10+' in str(val):
        return 10
    if '< 1' in str(val):
        return 0
    nums = re.findall(r'\d+', str(val))
    return int(nums[0]) if nums else np.nan

train['Years in current job'] = train['Years in current job'].apply(parse_years)
test['Years in current job']  = test['Years in current job'].apply(parse_years)

print("=== Years in current job после преобразования ===")
print(train['Years in current job'].value_counts().sort_index())
```

```python
#Объединение похожих категорий в Home Ownership
# 'Have Mortgage' фактически то же самое, что 'Home Mortgage'
train['Home Ownership'] = train['Home Ownership'].replace('Have Mortgage', 'Home Mortgage')
test['Home Ownership']  = test['Home Ownership'].replace('Have Mortgage', 'Home Mortgage')

print("=== Home Ownership после исправления ===")
print(train['Home Ownership'].value_counts())
```

```python
#Заполнение пропусков
#В этих признаках есть сильные выбросы — они тянут среднее вверх, делая его нерепрезентативным.
#Медиана устойчива к выбросам и лучше отражает "типичного" заёмщика.
# --- Числовые признаки ---
# Annual Income и Credit Score: медиана (вычисляем по train, применяем к обоим)
for col in ['Annual Income', 'Credit Score', 'Current Loan Amount',
            'Bankruptcies', 'Years in current job']:
    median_val = train[col].median()
    train[col] = train[col].fillna(median_val)
    test[col]  = test[col].fillna(median_val)
    print(f"{col}: заполнено медианой = {median_val}")

# Months since last delinquent: 54% пропусков —
# отсутствие записи само по себе несёт смысл (просрочек не было)
# Заполняем -1 как специальную метку
train['Months since last delinquent'] = train['Months since last delinquent'].fillna(-1)
test['Months since last delinquent']  = test['Months since last delinquent'].fillna(-1)
print("\nMonths since last delinquent: пропуски заполнены -1 (не было просрочек)")
```

```python
#Кодирование категориальных признаков
from sklearn.preprocessing import LabelEncoder

# Term: бинарный признак
train['Term'] = (train['Term'] == 'Long Term').astype(int)
test['Term']  = (test['Term'] == 'Long Term').astype(int)

# Home Ownership и Purpose: Label Encoding
for col in ['Home Ownership', 'Purpose']:
    le = LabelEncoder()
    le.fit(train[col])  # обучаем только на train!
    train[col] = le.transform(train[col])
    test[col]  = le.transform(test[col])
    print(f"{col}: закодирован")

print("\n=== Типы данных после кодирования ===")
print(train.dtypes)
```

```python
#Проверка: нет ли оставшихся пропусков
print("=== Пропуски в train после обработки ===")
print(train.isnull().sum())

print("\n=== Пропуски в test после обработки ===")
print(test.isnull().sum())
```

```python
#Масштабирование числовых признаков
from sklearn.preprocessing import StandardScaler

# Признаки для масштабирования (все числовые кроме целевой переменной)
feature_cols = [col for col in train.columns if col != 'Credit Default']

X_train = train[feature_cols].copy()
y_train = train['Credit Default'].copy()
X_test  = test[feature_cols].copy()

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   # fit только на train!
X_test_scaled  = scaler.transform(X_test)

# Сохраняем в DataFrame для удобства
X_train_scaled = pd.DataFrame(X_train_scaled, columns=feature_cols)
X_test_scaled  = pd.DataFrame(X_test_scaled,  columns=feature_cols)

print("=== Масштабирование завершено ===")
print(X_train_scaled.describe().T[['mean', 'std']].head(10))
```

```python
#Итоговая проверка
print(f"X_train shape: {X_train_scaled.shape}")
print(f"X_test shape:  {X_test_scaled.shape}")
print(f"y_train distribution:\n{y_train.value_counts()}")
```

**Ключевые решения этого шага**

* `Credit Score > 850` -> Делим на 10
* `Current Loan Amount = 99999999` -> Заменяем → медиана
* `Years in current job` — строки -> Преобразуем в числа `(0–10)`
* `Have Mortgage` -> Объединяем с `Home Mortgage`
* `Annual Income`, `Credit Score` — 20% пропусков -> Медиана по train
* `Months since last delinquent` — 54% пропусков -> Заполняем `-1` (смысловая метка)
* `Term` -> Бинаризируем `(0/1)`
* `Home Ownership`, `Purpose Label` -> Encoding
* `Все числовые признаки` -> StandardScaler

**Ключевые итоги шага**

* `Credit Score` исправлен — среднее упало с `1151` до `720`
* `Current Loan Amount` — 870 заглушек 99999999 заменено на `NaN` → медиана 265,826
* Пропусков не осталось — ни в `train`, ни в `test`
* Масштабирование — `mean ≈ 0`, `std ≈ 1` у всех признаков

```python
#Самописная логистическая регрессия
class LogisticRegressionCustom:
    def __init__(self, lr=0.1, n_iter=1000):
        self.lr = lr          # learning rate
        self.n_iter = n_iter  # число итераций градиентного спуска
        self.weights = None
        self.bias = None
        self.losses = []

    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y)
        n_samples, n_features = X.shape

        # Инициализация весов нулями
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.n_iter):
            # Прямой проход
            z = X @ self.weights + self.bias
            y_pred = self._sigmoid(z)

            # Binary Cross-Entropy Loss
            loss = -np.mean(y * np.log(y_pred + 1e-9) +
                            (1 - y) * np.log(1 - y_pred + 1e-9))
            self.losses.append(loss)

            # Градиенты
            error = y_pred - y
            dw = (X.T @ error) / n_samples
            db = np.mean(error)

            # Обновление весов
            self.weights -= self.lr * dw
            self.bias    -= self.lr * db

    def predict_proba(self, X):
        z = np.array(X) @ self.weights + self.bias
        return self._sigmoid(z)

    def predict(self, X, threshold=0.5):
        return (self.predict_proba(X) >= threshold).astype(int)
```

```python
#Подбор порога для самописной модели
from sklearn.metrics import f1_score

custom_lr = LogisticRegressionCustom(lr=0.1, n_iter=2000)
custom_lr.fit(X_tr.values, y_tr.values)

# Подбираем оптимальный порог по F1
proba_val = custom_lr.predict_proba(X_val.values)
thresholds = np.arange(0.1, 0.9, 0.01)
f1_scores  = [f1_score(y_val, (proba_val >= t).astype(int)) for t in thresholds]

best_threshold = thresholds[np.argmax(f1_scores)]
best_f1_custom = max(f1_scores)

print(f"Лучший порог:      {best_threshold:.2f}")
print(f"Лучший F1 (custom): {best_f1_custom:.4f}")

# График сходимости функции потерь
plt.figure(figsize=(10, 4))
plt.plot(custom_lr.losses)
plt.title('Сходимость функции потерь (Custom Logistic Regression)')
plt.xlabel('Итерация')
plt.ylabel('Binary Cross-Entropy Loss')
plt.tight_layout()
plt.show()
```

**Нужно добавить Feature Engineering.**

Новые фичи:
1. Долговая нагрузка: ежемесячный долг / месячный доход
2. Утилизация кредитного лимита

```python
#Feature Engineering на немасштабированных данных
# Создаём новые признаки из оригинальных данных (до масштабирования)
# train и test у нас уже обработаны (пропуски заполнены, категории закодированы)

fe_cols_source = train.drop(columns=['Credit Default'])

for df_orig, df_scaled in [(train.drop(columns=['Credit Default']), X_train_full),
                            (test, X_test_full)]:

    # 1. Долговая нагрузка: ежемесячный долг / месячный доход
    df_scaled['Debt to Income'] = (
        df_orig['Monthly Debt'] / (df_orig['Annual Income'] / 12 + 1)
    ).values

    # 2. Утилизация кредитного лимита
    df_scaled['Credit Utilization'] = (
        df_orig['Current Credit Balance'] / (df_orig['Maximum Open Credit'] + 1)
    ).values

    # 3. Кредит относительно дохода
    df_scaled['Loan to Income'] = (
        df_orig['Current Loan Amount'] / (df_orig['Annual Income'] + 1)
    ).values

    # 4. Флаг: была ли просрочка (-1 означает "не было")
    df_scaled['Has Delinquent'] = (
        (df_orig['Months since last delinquent'] != -1).astype(int)
    ).values

    # 5. Флаг: есть ли кредитные проблемы
    df_scaled['Has Credit Problems'] = (
        (df_orig['Number of Credit Problems'] > 0).astype(int)
    ).values

print(f"X_train_full shape: {X_train_full.shape}")
print(f"X_test_full shape:  {X_test_full.shape}")
print(f"\nНовые признаки: {X_train_full.columns[-5:].tolist()}")
```

```python
#Переобучаем с Feature Engineering
X_tr_fe, X_val_fe, y_tr_fe, y_val_fe = train_test_split(
    X_train_full, y_train,
    test_size=0.2, random_state=42, stratify=y_train
)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_fe = GridSearchCV(
    LogisticRegression(class_weight='balanced', random_state=42, max_iter=1000),
    {'C': [0.001, 0.01, 0.1, 1, 10, 100], 'solver': ['lbfgs', 'liblinear']},
    cv=cv, scoring='f1', n_jobs=-1
)
grid_fe.fit(X_train_full, y_train)

best_lr_fe = grid_fe.best_estimator_
y_pred_fe  = best_lr_fe.predict(X_val_fe)
f1_fe      = f1_score(y_val_fe, y_pred_fe)

print(f"Лучшие параметры: {grid_fe.best_params_}")
print(f"\n=== LogReg + Feature Engineering ===")
print(classification_report(y_val_fe, y_pred_fe,
      target_names=['0 — Нет дефолта', '1 — Дефолт']))
print(f"F1-score (класс 1): {f1_fe:.4f}")
```

```python
#Итоговая таблица всех моделей
print("\n========== ИТОГОВОЕ СРАВНЕНИЕ ВСЕХ МОДЕЛЕЙ ==========")
print(f"{'Модель':<50} {'F1-score':>10}")
print("-" * 62)
print(f"{'Baseline LogReg (7 признаков, sklearn)':<50} {f1_baseline:>10.4f}")
print(f"{'Tuned LogReg (все признаки, sklearn)':<50} {f1_full:>10.4f}")
print(f"{'Custom LogReg (все признаки)':<50} {best_f1_cust:>10.4f}")
print(f"{'Tuned LogReg + Feature Engineering (sklearn)':<50} {f1_fe:>10.4f}")
print(f"{'Custom LogReg + Feature Engineering':<50} {best_f1_cfe:>10.4f}")
print("-" * 62)

best_f1 = max(f1_fe, best_f1_cfe)
print(f"Лучший F1: {best_f1:.4f}")
print(f"Целевой порог F1 > 0.5: {'✅ ДОСТИГНУТ' if best_f1 > 0.5 else '❌ НЕ ДОСТИГНУТ'}")
```

Логистическая регрессия достигла своего потолка — ~0.485. Это фундаментальное ограничение линейной модели на нелинейных данных.

Решение одно — нужно добавить нелинейное преобразование признаков прямо внутри логистической регрессии через PolynomialFeatures. Это позволит модели улавливать взаимодействия между признаками.

```python
#Polynomial Features + LogReg
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

# Pipeline: полиномиальные признаки степени 2 → логистическая регрессия
pipe = Pipeline([
    ('poly', PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
    ('lr',   LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42))
])

param_grid_poly = {
    'lr__C':      [0.01, 0.1, 1, 10],
    'lr__solver': ['lbfgs', 'liblinear']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_poly = GridSearchCV(
    pipe, param_grid_poly,
    cv=cv, scoring='f1',
    n_jobs=-1, verbose=1
)
grid_poly.fit(X_train_full, y_train)

print(f"Лучшие параметры: {grid_poly.best_params_}")
print(f"Лучший F1 (CV):   {grid_poly.best_score_:.4f}")
```

```python
#Оцениваем Polynomial LogReg на валидации
X_tr_p, X_val_p, y_tr_p, y_val_p = train_test_split(
    X_train_full, y_train,
    test_size=0.2, random_state=42, stratify=y_train
)

best_pipe = grid_poly.best_estimator_
y_pred_poly = best_pipe.predict(X_val_p)
f1_poly = f1_score(y_val_p, y_pred_poly)

print("=== Polynomial LogReg ===")
print(classification_report(y_val_p, y_pred_poly,
      target_names=['0 — Нет дефолта', '1 — Дефолт']))
print(f"F1-score (класс 1): {f1_poly:.4f}")
```

```python
#Итоговая таблица + финальный прогноз
print("\n========== ИТОГОВОЕ СРАВНЕНИЕ ВСЕХ МОДЕЛЕЙ ==========")
print(f"{'Модель':<50} {'F1-score':>10}")
print("-" * 62)
print(f"{'Baseline LogReg (7 признаков)':<50} {f1_baseline:>10.4f}")
print(f"{'Tuned LogReg (все признаки)':<50} {f1_full:>10.4f}")
print(f"{'Custom LogReg (все признаки)':<50} {best_f1_cust:>10.4f}")
print(f"{'LogReg + Feature Engineering':<50} {f1_fe:>10.4f}")
print(f"{'Polynomial LogReg (degree=2)':<50} {f1_poly:>10.4f}")
print("-" * 62)

best_f1_all = max(f1_baseline, f1_full, best_f1_cust, f1_fe, f1_poly)
print(f"Лучший F1: {best_f1_all:.4f}")
print(f"Целевой порог F1 > 0.5: {'✅ ДОСТИГНУТ' if best_f1_all > 0.5 else '❌ НЕ ДОСТИГНУТ'}")

# Финальный прогноз лучшей моделью
y_test_poly = best_pipe.predict(X_test_full)
submission = pd.DataFrame({'Credit Default': y_test_poly})
submission.to_csv('predictions.csv', index=False)
print(f"\nПрогнозы сохранены ✅")
print(f"0 — {(y_test_poly==0).sum()}, 1 — {(y_test_poly==1).sum()}")
```
